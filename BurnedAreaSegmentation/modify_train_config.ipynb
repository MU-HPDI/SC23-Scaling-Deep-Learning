```python
training_options = {'lr':[0.0001,0.00001],
                    'batch_size':[8,16],
                    'optimizer':['Lamb']}
```

```python
folders_list = dict1.keys()
folders_list
```

# Create exp directory for each experiment

```python
source_dir = 'experiments_2'
if os.path.exists('{}'.format(source_dir)):
    print('{} alreay exists'.format(source_dir))
else:
    os.mkdir('{}'.format(source_dir))
```

```python
exp_list = []

for idx in list(range(8)):
    exp_list.append('exp{}'.format(idx+1))
exp_list
```

```python
for exp in exp_list:
    if os.path.exists('{}/{}'.format(source_dir,exp)):
        print('{}/{} alreay exists'.format(source_dir,exp))
    else:
        os.mkdir('{}/{}'.format(source_dir,exp))
```

# create output directory for each case in each experiment directory

```python
for exp in exp_list:
    for folder in folders_list:
        # output)dir
        output_dir = dict1[folder][0]
        if os.path.exists('{}/{}/{}'.format(source_dir,exp,output_dir)):
            print('{}/{}/{} already exists'.format(source_dir,exp,output_dir))
        else:
            os.mkdir('{}/{}/{}'.format(source_dir,exp,output_dir))
        
        if os.path.exists('{}/{}/{}/last'.format(source_dir,exp,output_dir)):
            print('last dire already exists')
        else:
            os.mkdir('{}/{}/{}/last'.format(source_dir,exp,output_dir))
            
        if os.path.exists('{}/{}/{}/lowest'.format(source_dir,exp,output_dir)):
            print('last dire already exists')
        else:
            os.mkdir('{}/{}/{}/lowest'.format(source_dir,exp,output_dir))
```

# copy config folder to each experiment directory

```python
for exp in exp_list:
    source = 'configs2'
    distination = '{}/{}/configs'.format(source_dir,exp)
    if os.path.exists(distination):
        print('{} exists'.format(distination))
    else:
        shutil.copytree(source,distination)
```

# Modify train config

```python
exp_rank = 0
for optimizer in training_options['optimizer']:
    for batch in training_options['batch_size']:
        for lr in training_options['lr']:
            
            
            if exp_rank > 19:
                #exp_rank = exp_rank + 1
                '''
                if exp_rank%3 == 1:
                    print()
                '''
                break
            else:
                exp_rank = exp_rank + 1
                print('{:2} {:6} {:3} {:6}'.format(exp_rank,lr,batch*2,optimizer))
                for folder in folders_list:
                    data_type = folder.split('_')[1]
                    # train config file
                    train_config_file = dict1[folder][1]
                    print('procesing train config file: {}/exp{}/configs/{}'.format(source_dir,exp_rank,train_config_file))
                    filein1 = open('{}/exp{}/configs/{}'.format(source_dir,exp_rank,train_config_file))
                    config = json.load(filein1)
                    filein1.close()

                    config['batch_size_per_gpu'] = batch
                    
                    config['optimizer_name']     = optimizer
                    config['optimizer_args']     = {"lr":lr}
                    
                    config['lr_scheduler']       = 'StepLR'
                    config['scheduler_args']     = {"step_size":50,"gamma":0.5}
                    
                    config['validate_while_training'] = True
                    config['training_data']      = [{"image_dir": "/canada2019-3/training_data_png/{}s/train".format(data_type),
                                                     "mask_dir": "/canada2019-3/training_data_png/labs/train"
                                                   }] 
                    config['validation_data']    = [{"image_dir": "/canada2019-3/training_data_png/{}s/validate".format(data_type),
                                                     "mask_dir": "/canada2019-3/training_data_png/labs/validate"
                                                   }] 

                    filein2 = open('{}/exp{}/configs/{}'.format(source_dir,exp_rank,train_config_file),'w')
                    json.dump(config,filein2, indent=4)
                    filein2.close()
                    
```

# Modify evaluate test data config

## config file to evaluate test data with last model

```python
exp_rank = 0
for optimizer in training_options['optimizer']:
    for batch in training_options['batch_size']:
        for lr in training_options['lr']:
            
            
            if exp_rank > 5:
                #exp_rank = exp_rank + 1
                '''
                if exp_rank%3 == 1:
                    print()
                '''
                break
            else:
                exp_rank = exp_rank + 1
                print('{:2} {:6} {:3} {:6}'.format(exp_rank,lr,batch*2,optimizer))
                
                for folder in folders_list:
                    data_type = folder.split('_')[1]
                    output_dir       = dict1[folder][0]
                    test_config_file = dict1[folder][2]
                    
                    if os.path.exists('{}/exp{}/{}/validation_metrics_history.csv'.format(source_dir,exp_rank,output_dir)):
                        valid_history = pd.read_csv('{}/exp{}/{}/validation_metrics_history.csv'.format(source_dir,exp_rank,output_dir))

                        loss_col = valid_history['loss']
                        last     = len(loss_col)

                        # test config file last model
                        print('procesing test config file: {}/exp{}/configs/{}'.format(source_dir,exp_rank,test_config_file))
                        filein3 = open('{}/exp{}/configs/{}'.format(source_dir,exp_rank,test_config_file))
                        config = json.load(filein3)
                        filein3.close()

                        model_current = config['pretrained_model_path']
                        model_last    = '/canada2019-3/{}/exp{}/{}/models/EPOCH{}.pt'.format(source_dir,exp_rank,output_dir,last)
                        config['pretrained_model_path'] = model_last

                        filein4 = open('{}/exp{}/configs/{}_last.json'.format(source_dir,exp_rank,test_config_file.split('.')[0]),'w')
                        json.dump(config,filein4, indent=4)
                        filein4.close()
                        
                    else:
                        print('    this model is not yet trained')
                        continue
```

## config file to evaluate test data with lowest validation loss model

```python
exp_rank = 0
for optimizer in training_options['optimizer']:
    for batch in training_options['batch_size']:
        for lr in training_options['lr']:
            
            
            if exp_rank > 5:
                #exp_rank = exp_rank + 1
                '''
                if exp_rank%3 == 1:
                    print()
                '''
                break
            else:
                exp_rank = exp_rank + 1
                print('{:2} {:6} {:3} {:6}'.format(exp_rank,lr,batch*2,optimizer))
                
                for folder in folders_list:
                    data_type = folder.split('_')[1]
                    output_dir       = dict1[folder][0]
                    test_config_file = dict1[folder][2]
                    
                    if os.path.exists('{}/exp{}/{}/validation_metrics_history.csv'.format(source_dir,exp_rank,output_dir)):
                        valid_history = pd.read_csv('{}/exp{}/{}/validation_metrics_history.csv'.format(source_dir,exp_rank,output_dir))

                        loss_col = valid_history['loss']
                        lowest   = loss_col.argmin()

                        # test config file lowest valid loss model
                        print('procesing test config file: {}/exp{}/configs/{}'.format(source_dir,exp_rank,test_config_file))
                        filein5 = open('{}/exp{}/configs/{}'.format(source_dir,exp_rank,test_config_file))
                        config = json.load(filein5)
                        filein5.close()

                        model_current = config['pretrained_model_path']
                        model_lowest   = '/canada2019-3/{}/exp{}/{}/models/EPOCH{}.pt'.format(source_dir,exp_rank,output_dir,lowest)
                        config['pretrained_model_path'] = model_lowest

                        filein6 = open('{}/exp{}/configs/{}_lowest.json'.format(source_dir,exp_rank,test_config_file.split('.')[0]),'w')
                        json.dump(config,filein6, indent=4)
                        filein6.close()
                        
                    else:
                        print('    this model is not yet trained')
                        continue
```
